```python
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import numpy as np
import alpaca_trade_api as tradeapi

%matplotlib inline
```

```python
load_dotenv()
```

```python
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")
```

```python
alpaca_secret_key
```

```python
alpaca_api_key
```

```python
alpaca = tradeapi.REST(alpaca_api_key, alpaca_secret_key, api_version="v2")
```

```python
alpaca
```

```python
today = pd.Timestamp("2023-06-30", tz="America/New_York").isoformat()
start = pd.Timestamp("2016-06-30", tz="America/New_York").isoformat()
start
```

```python
tickers=["ACWI", "IWV", "AGG", "NVDA", "TSLA", "AAPL"]
```

```python
timeframe = "1Day"
```

```python
portfolio_df = alpaca.get_bars(tickers, timeframe, start=start, end=today).df
portfolio_df
```

```python
ACWI = portfolio_df[portfolio_df["symbol"]=="ACWI"].drop("symbol", axis=1)
IWV = portfolio_df[portfolio_df["symbol"]=="IWV"].drop("symbol", axis=1)
AGG = portfolio_df[portfolio_df["symbol"]=="AGG"].drop("symbol", axis=1)

NVDA = portfolio_df[portfolio_df["symbol"]=="NVDA"].drop("symbol", axis=1)
TSLA = portfolio_df[portfolio_df["symbol"]=="TSLA"].drop("symbol", axis=1)
AAPL = portfolio_df[portfolio_df["symbol"]=="AAPL"].drop("symbol", axis=1)
```

```python
TSLA
```

```python
IWV
```

```python
AGG
```

```python
portfolio_df_all = pd.concat([ACWI, IWV, AGG, NVDA, TSLA, AAPL], axis=1, keys=["ACWI", "IWV", "AGG", "NVDA", "TSLA", "AAPL"])
portfolio_df_all
```

```python
portfolio_df_all_close = pd.DataFrame()
portfolio_df_all_close["ACWI"] = portfolio_df_all["ACWI"]["close"]
portfolio_df_all_close["IWV"] = portfolio_df_all["IWV"]["close"]
portfolio_df_all_close["AGG"] = portfolio_df_all["AGG"]["close"]

portfolio_df_all_close["NVDA"] = portfolio_df_all["NVDA"]["close"]
portfolio_df_all_close["TSLA"] = portfolio_df_all["TSLA"]["close"]
portfolio_df_all_close["AAPL"] = portfolio_df_all["AAPL"]["close"]

portfolio_df_all_close.index = portfolio_df_all_close.index.date
portfolio_df_all_close.head(10)
```

```python
portfolio_df_all_close.tail(10)
```

```python
df_daily_returns = portfolio_df_all_close.pct_change().dropna()
df_daily_returns.head()
```

```python
df_daily_returns.tail()
```

```python
us_equity_wt = 0.7
global_equity_wt = 0.2
fixed_inc_wt = 0.1

df_daily_returns["Portfolio"] = ((us_equity_wt*df_daily_returns["IWV"]) 
                                 + (global_equity_wt*df_daily_returns["ACWI"]) 
                                 + (fixed_inc_wt*df_daily_returns["AGG"]))
df_daily_returns
```

```python
df_cumulative = (1+df_daily_returns).cumprod()
```

```python
df_cumulative
```

```python
final_port_total_return = (df_cumulative.iloc[-1]['Portfolio'])-1
final_port_total_return
```

```python
ann_return = ((1+final_port_total_return)**(1/7))-1
ann_return
```

```python
sharpe_ratios = (df_daily_returns.mean() * 252) / (df_daily_returns.std() * np.sqrt(252))
sharpe_ratios.sort_values(ascending=False)
```

```python
rel_sharpe_ratios = sharpe_ratios - 1
```

```python
rel_sharpe_ratios.sort_values(ascending=False)
```

```python
df_daily_returns.iloc[:, 0:3].plot(figsize=(20, 10), kind="box", title="Daily Returns")
```
